```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import pandas as pd
import time
import re
import xlsxwriter
from googletrans import Translator

now =time.gmtime(time.time())
columns = ["title", "korean", "length_of_news", "link"]
# https://github.com/ssut/py-googletrans
google_translator = Translator(service_urls=[
        'translate.google.cn',
        'translate.google.com'
    ])
length_limit = 1000
```

```python
html_shine_metro_scroll = urlopen("https://www.shine.cn/news/metro")  

bs_shine_metro_scroll = BeautifulSoup(html_shine_metro_scroll, "html.parser") 

shine_metro_titles = []
shine_metro_translated_titles = []
shine_metro_links = []
shine_metro_lens = []

for link in bs_shine_metro_scroll.find_all("div", {"class": "list-story"}):

    html_shine_metro_title = link.find("h2").text.strip()
    html_shine_metro_link = link.get("href")
    if html_shine_metro_link.count("http"):
        shine_metro_titles.append(html_shine_metro_title)
#         print(html_shine_metro_title)
#         shine_metro_translated_titles.append(google_translator.translate(html_shine_metro_title, dest='ko').text)
        shine_metro_links.append(html_shine_metro_link)
print(shine_metro_titles)
```

```python
for html_online_sh_title, html_online_sh_link in zip(online_sh_titles, online_sh_links):
    html_online_sh = urlopen(html_online_sh_link)
    if html_online_sh_link.count("edu.sina.com.cn"):
        key = "article"
    else:
        key = "article-body main-body"
    bs_online_sh = BeautifulSoup(html_online_sh, "html.parser", from_encoding="cp936")
    content = ""
    try:
        for con in bs_online_sh.find('div', {'class' : key}).find_all('p'):
            temp_con = con.text.strip()
            if re.search(u'[\u4e00-\u9fff]', temp_con):
                content += temp_con
    except:
        pass
    online_sh_lens.append(len(content))
    

online_sh_data = {
    columns[0]:online_sh_titles,
    columns[1]:online_sh_translated_titles,
    columns[2]:online_sh_lens,
    columns[3]:online_sh_links
}

online_sh_frame = pd.DataFrame(online_sh_data, columns=columns)
online_sh_frame.sort_values(columns[2], axis = 0, ascending = True, inplace = True, na_position ='first') 
```

```python
#renminwang_edu
html_renmin_edu_scroll = urlopen("https://hot.online.sh.cn/node/node_65634.htm")  

bs_renmin_edu_scroll = BeautifulSoup(html_renmin_edu_scroll, "html.parser", from_encoding="cp936") 

renmin_edu_titles = []
renmin_edu_translated_titles = []
renmin_edu_links = []
renmin_edu_lens = []


for link in bs_renmin_edu_scroll.find_all("div", {"class": "list_thread"}):
    temp_link= link.find("a")
    html_renmin_edu_title = temp_link.text.strip()
    html_renmin_edu_link = temp_link.get("href")
#     print(html_renmin_edu_title, html_renmin_edu_link)
    if html_renmin_edu_link.count(str(now.tm_mday)):
        renmin_edu_titles.append(html_renmin_edu_title)
#         print(html_renmin_edu_title)
        renmin_edu_translated_titles.append(google_translator.translate(html_renmin_edu_title, dest='ko').text)
        renmin_edu_links.append(html_renmin_edu_link)
print(renmin_edu_titles)
print(renmin_edu_links)
```
